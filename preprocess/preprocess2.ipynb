## 소음진동민원전처리
소음진동민원 전처리 (2022,21,20년도)

- 공장, 교통 부분에 있는 (-) 를 0 으로
- 모든 값 int 로 통일
- 자치구별(1) 제거, 소계 행 버리기
- 연도칼럼 추가 (2022, 2021, 2020)

```python
import pandas as pd

df = pd.read_excel('./소음진동민원+현황_20240603212535.xlsx')
```

```python
df
```

```python
# 자치구별(1) 제거
df = df.drop(columns=['자치구별(1)'])
```

```python
# 소계 행 제거
df = df[df['자치구별(2)'] != '소계']
```

```python
df = df.replace('-', 0)
```

```python
df
```

```python
# 행의 모든 년도를 2020, 2021, 2022 로 변경
df.columns = ['자치구별(2)','2020','2020','2020','2020','2021','2021','2021','2021','2022','2022','2022','2022']
```

```python
df.drop([0], axis=0, inplace=True)
```

```python
df
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.fillna('')
df.columns = df.columns.tolist()
df.columns = ['자치구별', '2020_소계', '2020_공장', '2020_교통', '2020_생활', '2021_소계', '2021_공장', '2021_교통', '2021_생활', '2022_소계', '2022_공장', '2022_교통', '2022_생활']

melted_df = pd.melt(df, id_vars=['자치구별'], value_vars=['2020_소계', '2020_공장', '2020_교통', '2020_생활', '2021_소계', '2021_공장', '2021_교통', '2021_생활', '2022_소계', '2022_공장', '2022_교통', '2022_생활'],
                    var_name='Year_Type', value_name='Count')

melted_df['Year'] = melted_df['Year_Type'].apply(lambda x: x.split('_')[0])
melted_df['Noise_Type'] = melted_df['Year_Type'].apply(lambda x: x.split('_')[1])

final_df = melted_df[['Year', '자치구별', 'Noise_Type', 'Count']].reset_index(drop=True)

final_df.to_csv("./소음진동민원_전처리_최종", encoding='utf-8-sig', index=False)

final_df.head()
```

```python
# 불필요한 행 제거
final_df.drop([0], axis=0, inplace=True)
```

```python
filtered_df = final_df[final_df['자치구별'] != '자치구별(2)']

filtered_df.head()
```

```python
filtered_df.rename(columns={'자치구별': 'Gu'}, inplace=True)
filtered_df.rename(columns={'Noise_Type': 'NoiseType'}, inplace=True)
```

```python
filtered_df['Count'] = filtered_df['Count'].astype(int)

filtered_df.dtypes
```

```python
filtered_df.info()
```

```python
# 최종 데이터프레임 저장
restructured_data_path = "./최종_소음진동민원_전처리.csv"
filtered_df.to_csv(restructured_data_path, encoding='utf-8-sig', index=False)
```

## 서울시 환경오염물질 배출 시설 통계
- (-) 를 0 으로
- 모든 값 int 로 통일
- 지역별(1) 제거, 소계 행 버리기
- 대기, 수질 부분 → 대기 1종, 대기 2종 이런 통일
- 소음및진동 열 제거 → 설정에서 변경해서 다운받기
- 연도칼럼 추가 (2022, 2021, 2020)

```python
import pandas as pd

df = pd.read_excel('./환경오염물질+배출시설_20240603164633.xlsx')
```

```python
# 자치구별(1) 제거
df = df.drop(columns=['지역별(1)'])
```

```python
# 소계 행 제거
df = df[df['지역별(2)'] != '소계']
```

```python
df = df.replace('-', 0)
```

```python
df
```

```python
# 대기 1종, 대기 2종 이렇게 통일


# 두 번째와 세 번째 행을 결합하여 새로운 첫 번째 행 생성
new_first_row = df.iloc[0].astype(str) + " " + df.iloc[1].astype(str)

# 첫 번째 행으로 삽입하고, 기존 두 번째 및 세 번째 행 삭제
df.iloc[0] = new_first_row
df = df.drop([1])

# 인덱스 재설정
df.reset_index(drop=True, inplace=True)

df
```

```python
# 열 이름을 연도별로 정리하는 함수
def simplify_year(column_name):
    if column_name.startswith("2020"):
        return "2020"
    elif column_name.startswith("2021"):
        return "2021"
    elif column_name.startswith("2022"):
        return "2022"
    else:
        return column_name

# 열 이름 변환 적용
new_columns = [simplify_year(col) for col in df.columns]
df.columns = new_columns
```

```python
df.columns.values[0] = 'Gu'
```

```python
df.to_csv("./temp_file.csv", encoding='utf-8-sig', index=False)
```

```python
def convert_columns_to_int_where_possible(df):
    for col in df.columns:
        # 각 열에 대해 int로 변환 시도
        try:
            # int로 변환을 시도하고 성공하면 해당 열을 int로 변환
            if df[col].astype(int).equals(df[col].astype(int)):
                df[col] = df[col].astype(int)
        except ValueError:
            # int로 변환할 수 없는 경우, 열을 원래의 데이터 타입으로 유지
            continue
    return df


# 변환 함수 호출
final_df = convert_columns_to_int_where_possible(final_df)
```
