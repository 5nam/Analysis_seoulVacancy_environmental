## 스케일링에 따른 클러스터링 - standard / robust

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# 데이터 로드
file_path = './전체_데이터_합본.csv'
data = pd.read_csv(file_path)

# 소음, 대기 및 환경 조건 관련 열 선택
selected_columns = [
   '이산화질소농도(ppm)', '오존농도(ppm)', '일산화탄소농도(ppm)', '아황산가스농도(ppm)',
   '미세먼지농도(㎍/㎥)', '초미세먼지농도(㎍/㎥)', '소음_공장', '소음_교통', '소음_생활', '소음_소계',
   '대기 1종', '대기 2종', '대기 3종', '대기 4종', '대기 5종', '대기 소계', '수질 1종', '수질 2종',
   '수질 3종', '수질 4종', '수질 5종', '수질 소계'
]

data_for_clustering = data[selected_columns]

# Standard 스케일링
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(data_for_clustering)

# Robust 스케일링
robust_scaler = RobustScaler()
robust_scaled_data = robust_scaler.fit_transform(data_for_clustering)

# K-means 클러스터링 (StandardScaler)
kmeans_standard = KMeans(n_clusters=3, random_state=0)
clusters_standard = kmeans_standard.fit_predict(standard_scaled_data)

# K-means 클러스터링 (RobustScaler)
kmeans_robust = KMeans(n_clusters=3, random_state=0)
clusters_robust = kmeans_robust.fit_predict(robust_scaled_data)

# PCA를 사용한 2차원 데이터로의 차원 축소
pca = PCA(n_components=2)
principal_components_standard = pca.fit_transform(standard_scaled_data)
principal_components_robust = pca.fit_transform(robust_scaled_data)

# 클러스터링 결과 시각화 (StandardScaler)
plt.figure(figsize=(20, 8))

plt.subplot(1, 2, 1)
plt.scatter(principal_components_standard[:, 0], principal_components_standard[:, 1], c=clusters_standard, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('K-means Clustering (StandardScaler)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')

# 클러스터링 결과 시각화 (RobustScaler)
plt.subplot(1, 2, 2)
plt.scatter(principal_components_robust[:, 0], principal_components_robust[:, 1], c=clusters_robust, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('K-means Clustering (RobustScaler)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')

plt.tight_layout()
plt.show()
```

## 클러스터링 : 실루엣 분석, 엘보우 기법, 스케일링 적용 : 클러스터 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 데이터 로드 및 전처리
file_path = './전체_데이터_합본.csv'
data = pd.read_csv(file_path)

plt.rc('font', family='AppleGothic')

# 소음, 대기 및 환경 조건 관련 열 선택
selected_columns = ['이산화질소농도(ppm)', '소음_생활', '대기 1종', '수질 3종']
data_for_clustering = data[selected_columns]

# 스케일링
scaler = StandardScaler()
# scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# PCA를 사용한 2차원 데이터로의 차원 축소 및 시각화
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# 최적의 클러스터 수 찾기 (엘보우 기법)
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()

# 최적의 클러스터 수 찾기 (실루엣 분석)
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    clusters = kmeans.fit_predict(scaled_data)
    silhouette_scores.append(silhouette_score(scaled_data, clusters))

plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.show()


# K-means 클러스터링 (최적의 클러스터 수 선택 후)
optimal_k = 2  # 예를 들어 실루엣 분석 결과로 3을 선택했다고 가정
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=0)
clusters = kmeans.fit_predict(scaled_data)

plt.figure(figsize=(12, 8))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=clusters, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('K-means Clustering with Optimal Number of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

# 클러스터 별 빈집 평균 시각화 : 비율 차이

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 데이터 로드 및 전처리
file_path = './전체_데이터_합본.csv'
data = pd.read_csv(file_path)

plt.rc('font', family='AppleGothic')

# 소음, 대기 및 환경 조건 관련 열 선택
selected_columns = ['이산화질소농도(ppm)', '소음_생활', '대기 1종', '수질 3종']
data_for_clustering = data[selected_columns]

# 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# PCA를 사용한 2차원 데이터로의 차원 축소 및 시각화
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# 최적의 클러스터 수 선택
optimal_k = 4 
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=0)
clusters = kmeans.fit_predict(scaled_data)

# 클러스터 정보를 데이터프레임에 추가
data['Cluster'] = clusters


# 각 클러스터별 빈집 수의 평균 계산
average_vacant_houses = data.groupby('Cluster')['빈집_소계'].mean()
print(average_vacant_houses)

# 클러스터별 빈집 수 평균 간의 비율 계산
mean_difference = average_vacant_houses[1] - average_vacant_houses[2]
mean_ratio = (mean_difference / average_vacant_houses[2]) * 100
print(f"클러스터 1과 클러스터 2의 빈집 수 평균 차이 비율: {mean_ratio:.2f}%")

# 클러스터별 빈집 수 시각화 (평균 간의 차이 비율 표시)
plt.figure(figsize=(10, 6))
average_vacant_houses.plot(kind='bar', color=['purple','blue','pink','green'])
plt.title('클러스터별 빈집 수 평균')
plt.xlabel('클러스터')
plt.ylabel('빈집 수 평균')

# 평균 차이 표시
for i in range(len(average_vacant_houses)):
    plt.text(i, average_vacant_houses[i], f'{average_vacant_houses[i]:.2f}', ha='center', va='bottom')

plt.show()
```
